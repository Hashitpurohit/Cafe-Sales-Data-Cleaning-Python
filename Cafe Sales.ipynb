```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# reading dataset
df = pd.read_csv('/workspaces/IIM_Skills_Py_2026/data/Cafe Sales.csv')
```

```python
# data structure info
df.info()
```

```python
# dupicate records
print(f'Cafe Sales data contains {sum(df.duplicated())} duplicated record(s).')
```

```python
# missing records
df.isna().sum() * 100 / df.shape[0]
```

```python
# unique rcords
df_columns = set(df.columns)
exclude = {'transaction_id'}
df_columns = df_columns.difference(exclude)
```

```python
# unique records 
for column in df_columns:
    print(f'{column}:\n {df[column].unique()}', end = '\n\n')
```

```python
# replacing ERROR and UNKNOWN with NaN
df = df.replace(['ERROR','UNKNOWN'], np.nan)
```

```python
# isnpecting first few records
df.head()
```

```python
# fixing data types
df['Quantity'] = df['Quantity'].astype('Int64') 
df[['Price Per Unit', 'Total Spent']] = df[['Price Per Unit', 'Total Spent']].astype('float')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
# Dropping records where transaction date is missing
df = df.dropna(subset=['Transaction Date'])
```

```python
df.info()
```

```python
# replacing missinng values in payment method and location with INA flag
df[[ 'Payment Method', 'Location']] = df[['Payment Method', 'Location']]. fillna('INA')
```

```python
# replacing missing values in Item with mode value
df[ 'Item'] = df[ 'Item'].fillna(df['Item'].mode()[0])
```

```python
# replacinng missing values in quantity with median value
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
# inspecting price per unit by items
df.groupby('Item')['Price Per Unit'].value_counts()
```

```python
# filling missing values in price per unit based on mode value of respective items
df['Price Per Unit'] = df.groupby('Item')['Price Per Unit'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# filling missing values in total spent
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

```python
df.isna().sum() * 100 / df.shape[0]
```

```python
# saving cleaned data
df.to_csv('/workspaces/IIM_Skills_Py_2026/data/Cafe Sales(Cleaned).csv',index=False)
```
